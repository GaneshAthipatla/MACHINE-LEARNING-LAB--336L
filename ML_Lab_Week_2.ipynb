```python
#4)Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
import matplotlib.pyplot as plt

x_axis = [1, 2, 3, 4, 5]
y_axis = [2, 4, 6, 8, 10]

plt.plot(x_axis, y_axis)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")

plt.show()
```

```python
#5)write a python program to plot two or more lines with legends, different widths and colors
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 5, 7, 9, 11]
y3 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Line 1', color='blue', linewidth=2)
plt.plot(x, y2, label='Line 2', color='green', linewidth=3)
plt.plot(x, y3, label='Line 3', color='red', linewidth=1)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines with Legends")

plt.legend()

plt.show()
```

```python
#6)write a python program to plot quantities which have an x and y position
import matplotlib.pyplot as plt

data = [
    (1, 2),
    (3, 5),
    (5, 8),
    (7, 3),
    (9, 1),
]

x = [point[0] for point in data]
y = [point[1] for point in data]


plt.scatter(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Quantities")
plt.show()

```

```python
#7)create plot using subplots(), scatter, and some  other methods on the axes object
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

x = [1, 2, 3, 4, 5]
y1 = [3, 5, 7, 1, 4]
y2 = [2, 4, 6, 8, 10]


ax1.scatter(x, y1, marker='o', color='red', s=100)
ax1.grid(True, linestyle='--', linewidth=0.5)
ax1.set_title("Scatter Plot with Grid")


ax2.plot(x, y2, linestyle='--', color='green', linewidth=2)
ax2.set_xlim(0, 6)
ax2.set_ylim(0, 12)
ax2.set_title("Line Plot with Limits")

plt.tight_layout()
plt.show()
```

```python
#9)Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.
import matplotlib.pyplot as plt

product_names = ["Laptop", "Smartphone", "Smartwatch", "Headphones", "Fitness Tracker", "Gaming Console", "Smart Speaker"]

sales_data = [5200, 8100, 3500, 10000, 2800, 1200, 1500]

plt.bar(product_names, sales_data, color=['green', 'blue', 'red', 'purple', 'orange', 'pink', 'grey'])

plt.xlabel("Product Name")
plt.ylabel("Sales (Units)")
plt.title("Comparison of Product Sales")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#10)Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data)

plt.xlabel("Data Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")

plt.show()

```

```python
#11)Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.
import matplotlib.pyplot as plt

expenses = {
    "Housing": 500,
    "Food": 300,
    "Transportation": 200,
    "Utilities": 150,
    "Entertainment": 100,
    "Miscellaneous": 50
}

total_expense = sum(expenses.values())

percentages = {category: (amount / total_expense) * 100 for category, amount in expenses.items()}

plt.pie(percentages.values(), labels=percentages.keys(), autopct="%1.1f%%")

plt.title("Distribution of Expenses")

plt.show()
```
